### STEP 3: Frequency Analysis



#### Question: How often do earthquakes occur per year/month?

#### Insight:

###### Long quiet periods

###### Sudden spikes (seismic swarms)

#### 3.1 Monthly Frequency

```python
monthly_freq = (
    df_raw
    .groupby("month")
    .size()
    .reset_index(name="earthquake_count")
)

monthly_freq
```

```python
plt.figure(figsize=(8,5))
plt.bar(monthly_freq["month"], monthly_freq["earthquake_count"])
plt.title("Monthly Earthquake Frequency")
plt.xlabel("Month")
plt.ylabel("Number of Earthquakes")
plt.show()
```

##### Insight:

###### The highest earthquake frequency is observed in April and May, which aligns with the Pre-Monsoon peak.

###### Early-year months (January–March) and late-year months show moderate but consistent activity.

###### This pattern indicates a temporal clustering of events rather than random distribution, reinforcing the importance of time-based risk assessment.

#### EARTHQUAKE BY SEASONS 

```python

df_raw["season"] = np.select(
    [
        df_raw["month"].isin([12, 1, 2]),
        df_raw["month"].isin([3, 4, 5]),
        df_raw["month"].isin([6, 7, 8, 9]),
        df_raw["month"].isin([10, 11])
    ],
    [
        "Winter",
        "Pre-Monsoon",
        "Monsoon",
        "Post-Monsoon"
    ],
    default="Unknown"
)
```

```python
df_raw["season"].replace("Unknown", np.nan, inplace=True)
df_raw.head()
```

```python
seasonal_counts = (
    df_raw.groupby("season")
    .size()
    .sort_values(ascending=False)
)

seasonal_counts.plot(kind="bar")
plt.title("Earthquake Frequency by Season (Nepal Region)")
plt.ylabel("Number of Earthquakes")
plt.show()
```

#### 3.2 Seasonal Frequency

```python
seasonal_freq = (
    df_raw
    .groupby("season")
    .size()
    .reset_index(name="earthquake_count")
)

seasonal_freq
```

```python
plt.figure(figsize=(7,5))
plt.bar(seasonal_freq["season"], seasonal_freq["earthquake_count"])
plt.title("Earthquake Frequency by Season")
plt.xlabel("Season")
plt.ylabel("Number of Earthquakes")
plt.show()
```

##### Insight:

###### Earthquake occurrences are highest during the Pre-Monsoon season, followed by the Monsoon season.

###### Winter records the lowest number of earthquakes, indicating relatively lower seismic activity.

###### This seasonal variation suggests that while earthquakes are primarily tectonic, seasonal stress changes (such as temperature variation or hydrological loading) may influence the frequency of detectable events.

###### However, the presence of earthquakes across all seasons confirms that seismic activity in Nepal is persistent year-round.

#### 3.3 Frequency vs Time

```python
post_2015 = df_raw[df_raw["year"] >= 2015]

post_2015_freq = (
    post_2015
    .groupby("year")
    .size()
    .reset_index(name="earthquake_count")
)

post_2015_freq
```

```python
plt.figure(figsize=(8,5))
plt.plot(post_2015_freq["year"], post_2015_freq["earthquake_count"], marker="o")
plt.title("Post-2015 Earthquake Frequency")
plt.xlabel("Year")
plt.ylabel("Number of Earthquakes")
plt.grid(True)
plt.show()
```

##### Insight:

###### A sharp spike in 2015 reflects the major Nepal earthquake and its aftershock sequence.

###### In subsequent years, earthquake frequency declines significantly, indicating a post-seismic relaxation phase.

###### Minor fluctuations in later years are normal and expected in an active tectonic zone.

###### This confirms that 2015 is not a data error, but a genuine seismic anomaly.

##### “The dataset correctly captures post-mainshock aftershock decay.”

#### 5.4 Earthquake Density Grid

```python
df_raw["latitude"] = pd.to_numeric(df_raw["latitude"], errors="coerce")
df_raw["longitude"] = pd.to_numeric(df_raw["longitude"], errors="coerce")
df_raw.dtypes
df_raw = df_raw.dropna(subset=["latitude", "longitude"])
```

```python
lat_bins = pd.cut(df_raw["latitude"], bins=10)
lon_bins = pd.cut(df_raw["longitude"], bins=10)

density = (
    df_raw
    .groupby([lat_bins, lon_bins], observed=True)
    .size()
    .reset_index(name="count")
)

density.head()
```

```python
# Bin centers (for plotting)
density["lat_center"] = density["latitude"].apply(lambda x: x.mid)
density["lon_center"] = density["longitude"].apply(lambda x: x.mid)

# Plot
plt.figure()
plt.scatter(
    density["lon_center"],
    density["lat_center"],
    s=density["count"]
)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Earthquake Density Grid")
plt.show()
```

###### Each point = a geographical grid

###### Bigger circle = more earthquakes

###### Shows seismically active zones visually

###### Helps identify high-risk regions

#### 6.3 Depth Distribution by Number of Stations (nst)

#### 6.4 Depth vs Azimuthal Gap 

##### Lower gap → better location precision. Let’s see how depth affects location accuracy.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df_depth,
    x="depth",
    y="gap",
    hue="magnitude",
    palette="coolwarm",
    alpha=0.6
)
plt.xlabel("Depth (km)")
plt.ylabel("Azimuthal Gap (degrees)")
plt.title("Depth vs Azimuthal Gap")
plt.show()
```

#### Insight:

##### Most earthquakes occur at shallow depths (0–40 km).

##### Azimuthal gap values are mostly below 180°, indicating reasonably good station coverage.

##### Deeper earthquakes tend to show more scattered gap values, suggesting lower location reliability at higher depths.

##### Large-magnitude events are generally associated with moderate gap values, meaning they are better detected.

##### Small gaps (<180°) are good.
##### Deep earthquakes may have larger gaps if stations are sparse.

#### 6.5 Depth vs RMS (rms)

```python

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df_depth,
    x="depth",
    y="rms",
    hue="magnitude",
    palette="plasma",
    alpha=0.6
)
plt.xlabel("Depth (km)")
plt.ylabel("RMS (seconds)")
plt.title("Depth vs RMS Travel Time Residuals")
plt.show()
```

### STEP 7: Data Quality Analysis 



#### Question: Can we trust older data? / Did sensor networks improve?

#### Metrics:

###### gap

###### rms

###### horizontalError


#### 7.1 YEAR TREND OF GAP

```python
gap_trend = (
    df_depth.groupby("year")["gap"]
    .mean()
)

import matplotlib.pyplot as plt

plt.figure()
gap_trend.plot()
plt.xlabel("Year")
plt.ylabel("Average Azimuthal Gap (degrees)")
plt.title("Trend of Azimuthal Gap Over Time")
plt.grid(True)
plt.show()
```

#### Insight:

##### The azimuthal gap generally decreases over time, with fewer extreme gaps in recent years. Lower azimuthal gaps indicate better station distribution around epicenters, leading to more reliable earthquake location estimates.

##### Older earthquakes are less reliably located due to sparse station geometry.

#### 7.2 YEARLY RMS TREND

##### What is RMS?

##### Measures mismatch between observed and calculated seismic waves

##### Lower RMS = better location accuracy

```python
rms_trend = (
    df_depth.groupby("year")["rms"]
    .mean()
)

plt.figure()
rms_trend.plot()
plt.xlabel("Year")
plt.ylabel("Average RMS (seconds)")
plt.title("RMS Error Trend Over Time")
plt.grid(True)
plt.show()
```

#### INSIGHT
##### RMS travel-time residuals show a clear long-term decreasing trend, indicating improved seismic modeling, better station coverage, and more accurate arrival-time picking in recent years.

##### This strongly supports the idea that sensor networks and processing quality improved over time, especially after 2010.

#### 7.3 Horizontal Error (horizontalError)
##### What is Horizontal Error?

##### Estimated uncertainty in epicenter location (km)

##### Smaller values → more precise location

```python
h_error_trend = (
    df_depth.groupby("year")["horizontalError"]
    .mean()
)

plt.figure()
h_error_trend.plot()
plt.xlabel("Year")
plt.ylabel("Average Horizontal Error (km)")
plt.title("Horizontal Location Error Over Time")
plt.grid(True)
plt.show()
```

#### INSIGHT
##### The horizontal location error shows noticeable variation over time, with lower errors in the mid-2010s and a slight increase in recent years. This suggests improvements in epicenter location accuracy followed by variability likely due to changes in station availability or event distribution rather than true degradation in technology.

# ✅ Final Conclusion

### This project analyzed **35+ years of seismic activity in Nepal** using raw and processed earthquake records to understand **temporal patterns, spatial distribution, magnitude behavior, depth characteristics, and data quality trends**.

### Key conclusions:

##### **Seismic activity in Nepal is dominated by shallow earthquakes**, with most events occurring within **5–25 km depth**, consistent with active crustal tectonics along plate boundaries.

##### **Larger magnitude earthquakes are less frequent but better recorded**, showing stronger signals and higher station participation compared to smaller events.

##### **Seasonal and temporal analysis** indicates uneven earthquake reporting across years, with major spikes (e.g., 2015) reflecting **real seismic events rather than data errors**.

##### **Spatial density analysis** highlights earthquake clustering along known fault zones, confirming strong geographic patterns rather than random distribution.

##### **Data quality metrics (gap, rms, station count)** show clear improvement over time, indicating that **older records are less reliable**, and some observed trends may reflect **advancements in sensor networks rather than geological changes**.

##### **Deeper earthquakes tend to have **higher uncertainty**, fewer reporting stations, and larger azimuthal gaps, emphasizing the importance of quality filtering in seismic analysis.

### Overall takeaway:

#### > Nepal’s earthquake risk is primarily driven by frequent shallow seismic events, while improvements in seismic monitoring over time have significantly enhanced data reliability. Any long-term seismic trend analysis must carefully account for evolving sensor coverage and measurement accuracy.
